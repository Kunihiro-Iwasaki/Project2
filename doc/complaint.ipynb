### scrape by playwright(succeed)

```python
import random
import asyncio
import re
```

```python
from playwright.async_api import async_playwright

async def main():
    
    complaint_numbers = pd.read_csv('complaint_number.csv')

    results = []

    async with async_playwright() as p:

        browser = await p.chromium.launch(
            headless=False
        )

        page = await browser.new_page()

        for complaint_no in complaint_numbers['complaint_number']:
            url = (
            "https://a810-bisweb.nyc.gov/bisweb/"
            f"OverviewForComplaintServlet?complaintno={complaint_no}"
        )
            page = await browser.new_page()
            try:
                await page.goto(url, wait_until='domcontentloaded', timeout=60000)
                html = await page.content()
                await asyncio.sleep(random.uniform(3, 8))
                soup = BeautifulSoup(html, "html.parser")
            except:
                await page.close()
                continue

            result = {
                 'complaint_no': complaint_no,
                 'title': soup.title.get_text(strip=True),
                 'complaint_at': None,
                 'bin':None,
                 'owner':None,
                 'disposition':None,
                 'comments':None
                }
            
            # Complaint at
            tag = soup.find('td',string=re.compile('Complaint at:'))
            if tag:
                result['complaint_at'] = (
            tag.get_text(' ', strip=True)
            .replace('Complaint at:', '')
            .strip()
        )
            #bin
            bin_text = soup.find(string=re.compile('BIN:'))
            if bin_text:
                result['bin'] = (
                    bin_text.parent.get_text(' ', strip=True).replace('BIN:', '').strip()
                )

            #owner
            owner = soup.find(string=re.compile('^Owner:'))
            if owner:
                result['owner'] = (
                    owner.find_next('td').get_text(' ', strip=True)
                )

            #Disposition
            disp = soup.find(string=re.compile('^Disposition:'))
            if disp:
                result['disposition'] = (
                    disp.parent.find_next('td').get_text(' ', strip=True)
                )

            #commnets
            comments = soup.find(string=re.compile('^Comments:'))
            if comments:
                result['comments'] = (
                    comments.parent.find_next('td').get_text(' ', strip=True)
                    )
            results.append(result)

        await browser.close()
        
    df = pd.DataFrame(results)

    return df
 
df = await main()
```

```python
df.to_csv('complaint_detail.csv', index=False)
```
