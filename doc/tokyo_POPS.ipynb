```python
import pandas as pd
df = pd.read_csv('tokyo_POPS.csv')
df
```

```python
df['address'] = '東京都' + df['位置']
df
```

```python
import geocoder
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GEO_API_KEY"][:8] + "..."
```

```python
rows = []
for address in df['address']:
    g = geocoder.google(address, key=os.environ.get('GEO_API_KEY'))
    if g.latlng:
        rows.append(g.latlng)
    else:
        rows.append([None, None])
df['lat'] = [r[0] for r in rows]
df['lng'] = [r[1] for r in rows]
df
```

```python
rows = []
for i,address in enumerate(df['address'], start=1):
    try:
        print(f'{i}/{len(df)}{address}')
        g = geocoder.google(address, key=os.environ.get('GEO_API_KEY'), timeout=30)
        if g.latlng:
            rows.append(g.latlng)
        else:
            rows.append([None, None])
    except Exception as e:
        print('error', address)
        rows.append([None, None])

df['lat'] = [r[0] for r in rows]
df['lng'] = [r[1] for r in rows]
df
```
