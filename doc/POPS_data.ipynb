```python
import os
import requests
import pandas as pd
from sodapy import Socrata
from dotenv import load_dotenv
load_dotenv()
os.environ["API_KEY_1"][:8] + "..."
```

```python
#Privately Owned Public Spaces API endpoint
#url = 'https://data.cityofnewyork.us/api/v3/views/rvih-nhyn/query.json'
#response = requests.get(url)
#data = response.json()
#df = pd.DataFrame(data)
#df['building_address_with_zip']
```

```python
# Example authenticated client (needed for non-public datasets):
client = Socrata("data.cityofnewyork.us",
                  os.environ.get("API_KEY_1"),
                  username=os.environ.get("USER_NAME"),
                  password=os.environ.get("PASS")
                )

# First 1000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("rvih-nhyn", limit=1000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df
```

```python
results_df['building_address_with_zip'].value_counts()
```

```python
results_df['sf'] = (
    results_df['size_required'].str.findall(r'(\d+(?:\.\d+)?)\s*sf')
)
results_df['total_sf'] = results_df['sf'].apply(
    lambda x: sum(float(v) for v in x) if isinstance(x, list) else pd.NA
)
results_df
```

```python
results_df.columns.tolist()
```

```python
results_df2 = results_df[['building_name','pops_number','building_address_with_zip',
                          'year_completed','public_space_type','total_sf','size_required',
                          'hour_of_access_required','amenities_required','physically_disabled',
                          'developer','borough_name','latitude','longitude','bin','bbl']]
results_df2 = results_df2.sort_values('total_sf', ascending=False)
results_df2.head()
```

```python
results_df2['address'] = (
    results_df2['building_address_with_zip']
        .astype(str)
        .str.upper()
        .str.replace(r"[\s,]+", '', regex=True)
)
results_df2.head()
```

```python
results_df2['address_key'] = results_df2['address'].str.replace('NEWYORKNY', '', regex=False)
results_df2.to_csv('POPS.csv', index=False)
```

### Focus on violations

```python
target_code = ['A8', 'A1', 'A9', 'H3']
filtered_merged_df = merged_df[merged_df['disposition_code'].isin(target_code)]
filtered_merged_df
```

### Number of Buildings in Violation

```python
len(filtered_merged_df['bin'].unique())
```

### disposition date

```python
filtered_merged_df['disposition_date'] = pd.to_datetime(filtered_merged_df['disposition_date'])
filtered_merged_df.sort_values('disposition_date')['disposition_date'].tolist()
```

### Number of Violations by Building

```python
summary = (
    filtered_merged_df.groupby(
        ['bin']
        )['disposition_code']
    .value_counts()
    .unstack(fill_value=0)
    .reset_index()
)
summary['total_violation'] = (summary[['A1', 'A8', 'A9', 'H3']].sum(axis=1))
summary
```

### data cleaning

```python
building_info = (
    filtered_merged_df[
        [
            'bin',
            'building_name',
            'building_address_with_zip',
            'year_completed',
            'public_space_type',
            'hour_of_access_required',
            'size_required',
            'total_sf',
            'amenities_required',
            'physically_disabled',
            'developer',
            'complaint_number',
            'latitude',
            'longitude'
        ]
    ]
    .drop_duplicates('bin')
)

summary_final = pd.merge(summary,
    building_info,
    on='bin',
    how='left'
)

date_range = (
    filtered_merged_df.groupby('bin')['disposition_date']
    .agg(
        first_disposition_date='min',
        last_disposition_date='max'
    )
        .reset_index()
)

cleaned_summary = pd.merge(summary_final, date_range, on='bin', how='left')
cleaned_summary.to_csv('merged_POPS_DOB.csv', index=False)
cleaned_summary
```
